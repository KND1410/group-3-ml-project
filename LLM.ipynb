### Create and save your own API in .env with GEMINI_API=AIza..., remember to put .env in .gitignore

```python
from google import genai
from dotenv import load_dotenv
import os

load_dotenv()
client = genai.Client(api_key=os.getenv("GEMINI_API"))
uploadFile = client.files.upload(file="./IMDB_preprocessed20.csv")
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[uploadFile, "What is main content of this dataset about?"]
)

print(response.text)
```
