## Preprocessing

Removing the html tags

```python
import re

def remove_html(text):
    clean = re.sub(r'<.*?>', '', text)
    return clean

text = "Je suis <br> <hi guys> gay!!!"
print(remove_html(text))
```

Removing special characters (remove special characters, keep only letters)

```python
def remove_special_char(text):
    clean = re.sub(r'[^a-zA-Z\s]', '', text)
    return clean

text = "J'ai 10/10 envie de faire l'amour avec les enfants!!!!"
print(remove_special_char(text))
```

Tokenization (separate into words)

```python
def tokenize(text):
    token = text.split()
    return token

text = "Bonjour, je suis Khoi"
print(tokenize(text))
```

### Check point Testing

```python
def clean_text(text):
    text = remove_html(text)
    text = remove_special_char(text)
    text = text.lower()
    token = tokenize(text)
    return token

text = "Bonjour, je suis Khoi. Je suis <br> <hi guys> gay!!! Et j'ai 10/10 envie de faire l'amour avec les enfants!!!!"
print(clean_text(text))
```

## Normalization

Removing the stop words (Stopwords = common words that don't have a significant meaning.)

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_word = set(stopwords.words('english') + stopwords.words('french'))
print(stop_word)

def remove_stopword(text):
    filtered = [word for word in text if word not in stop_word]
    return filtered

text = ["bonjour", "je", "suis", "très", "gay"]
print(remove_stopword(text))
```

Lemmatization (Restore the word to its original form based on its true meaning)

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize(text):
    result = []
    for word in text:
        lemma = lemmatizer.lemmatize(word, pos='v')
        result.append(lemma)
    return result

text = ["kids", "playing", "toys"]
print(lemmatize(text))
```

### Checkpoint Testing

```python
def preprocessing(text):
    #Checkpoint 1
    text = remove_html(text)
    text = remove_special_char(text)
    text = text.lower()
    tokens = tokenize(text)
    
    #Checkpoint 2
    tokens = remove_stopword(tokens)
    tokens = lemmatize(tokens)
    return tokens

review = "This movie was <br />about slave AMAZING! I love watching nick R! The acting was the best I've ever seen. (The slave was being beaten by their owner)"
print(preprocessing(review))
```

## Vectorization

### Using Bag of Words

Count how many times each word appears.

```python
from sklearn.feature_extraction.text import CountVectorizer


reviews = [
    "movie amazing",
    "movie terrible boring",
    "amazing acting"
]

vectorizer = CountVectorizer()
x = vectorizer.fit_transform(reviews)
print(vectorizer.get_feature_names_out())
print(x.toarray())
```

### Using TF-IDF

One important thing to note: TfidfVectorizer accepts strings, not lists. Therefore, you need to join the tokens.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

reviews = [
    "movie amazing",
    "movie terrible boring",
    "amazing acting"
]

tfidf = TfidfVectorizer()
x = tfidf.fit_transform(reviews)
print(tfidf.get_feature_names_out())
print(x.toarray().round(2))
```

## Full pipeline

```python
reviews_raw = [
    "This movie was <br />AMAZING!!!",
    "Terrible film. I hated every minute."
]

review_clean = [' '.join(preprocessing(r)) for r in reviews_raw]
print(review_clean)

tfidf = TfidfVectorizer()
x = tfidf.fit_transform(review_clean)
print(tfidf.get_feature_names_out())
print(x.toarray().round(2))
```
