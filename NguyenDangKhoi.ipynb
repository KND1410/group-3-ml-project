```python
from google import genai
from dotenv import load_dotenv
import os

load_dotenv()
client = genai.Client(api_key=os.getenv("GEMINI_API")) #Create and save your own API in .env with GEMINI_API=AIza..., remember to put .env in .gitignore
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="I have the IMDB movie review dataset (50k reviews, binary sentiment: positive/negative). Suggest 5 specific features I could engineer from the raw review text to improve the model."
)

print(response.text)
```
